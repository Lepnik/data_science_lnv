```python
import pandas as pd
import plotly.express as px
solar_data = pd.read_csv('data/solarpower.csv')
#display(solar_data)

for i in range(len(solar_data['cum_power'])):
    if solar_data.iloc[i,1]==-1 and 0<i<len(solar_data['cum_power']):
        solar_data.iloc[i,1]=(solar_data.iloc[i-1,1]+solar_data.iloc[i+1,1])/2

solar_data["date"] = solar_data["date"].astype("datetime64[ns]")       
res = solar_data["date"].dt.to_period("M")

solar_data['day_power'] = solar_data['cum_power'].diff()
df = pd.DataFrame(solar_data['day_power'].groupby(res).mean())
display(df)
#solar_data['month']=
#solar_data['year']=
#solar_data['day_power ']=

fig = px.line(
    df, #датафрей
    #x='date', #ось абсцисс
    #y='day_power', #ось ординат
)
#fig.show()
```
