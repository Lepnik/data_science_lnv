```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn import model_selection
from sklearn import linear_model #линейные модели
from sklearn import metrics
from sklearn import ensemble
```

```python
dt = pd.read_csv('boston.csv')
dt.info()
```

```python
dt.head()
```

```python
dt['crim_rate'] = dt['crim_rate'].str.replace(',', '.').astype('float')
dt['zn'] = dt['zn'].str.replace(',', '.').astype('float')
dt['business'] = dt['business'].str.replace(',', '.').astype('float')
dt['nit_oxiden'] = dt['nit_oxiden'].str.replace(',', '.').astype('float')
dt['rooms'] = dt['rooms'].str.replace(',', '.').astype('float')
dt['age'] = dt['age'].str.replace(',', '.').astype('float')
dt['dist'] = dt['dist'].str.replace(',', '.').astype('float')
dt['pup_per_teaс'] = dt['pup_per_teaс'].str.replace(',', '.').astype('float')
dt['lower'] = dt['lower'].str.replace(',', '.').astype('float')
dt['target'] = dt['target'].str.replace(',', '.').astype('float')
dt.info()
```

```python
print(dt.corr()['target'].sort_values(ascending=False))
```

```python
X = dt.drop(['target'], axis=1)
y = dt['target']
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=13)#stratify=y, 
print('Train shape: {}'.format(X_train.shape))
print('Test shape: {}'.format(X_test.shape))
```

```python
round(sum(y_train)/len(y_train), 2)
```

```python
#Создаём объект класса LogisticRegression
lin_reg = linear_model.LinearRegression()
#Обучаем модель, минизируя logloss
lin_reg.fit(X_train, y_train)
y_pred_lin = lin_reg.predict(X_test)
#Выводим MAE
print(round(metrics.mean_absolute_error(y_test, y_pred_lin), 2))
```

```python
# Инициализируем модель дерева решений с максимальной глубиной 1 и обучаем её
dest = tree.DecisionTreeRegressor(random_state=13)
dest.fit(X_train, y_train)
y_pred_dest = dest.predict(X_test)
#Выводим результирующие коэффициенты
print(round(metrics.mean_absolute_error(y_test, y_pred_dest), 2))
```

```python
forest1 = ensemble.RandomForestRegressor(n_estimators=3, random_state = 13)
forest1.fit(X_train, y_train)
y_pred_forest1 = forest1.predict(X_test)
#Выводим результирующие коэффициенты
print(round(metrics.mean_absolute_error(y_test, y_pred_forest1), 2))
```

```python
forest2 = ensemble.RandomForestRegressor(n_estimators=10, random_state = 13)
forest2.fit(X_train, y_train)
y_pred_forest2 = forest2.predict(X_test)
#Выводим результирующие коэффициенты
print(round(metrics.mean_absolute_error(y_test, y_pred_forest2), 2))
```

```python
forest3 = ensemble.RandomForestRegressor(n_estimators=100, random_state = 13)
forest3.fit(X_train, y_train)
y_pred_forest3 = forest3.predict(X_test)
#Выводим результирующие коэффициенты
print(round(metrics.mean_absolute_error(y_test, y_pred_forest3), 2))
```

```python
forest4 = ensemble.RandomForestRegressor(n_estimators=500, random_state = 13)
forest4.fit(X_train, y_train)
y_pred_forest4 = forest4.predict(X_test)
#Выводим результирующие коэффициенты
print(round(metrics.mean_absolute_error(y_test, y_pred_forest4), 2))
```
