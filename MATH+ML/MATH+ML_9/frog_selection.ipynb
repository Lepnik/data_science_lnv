```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn import model_selection
from sklearn import linear_model #линейные модели
from sklearn import metrics
from sklearn import ensemble
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('Frogs_MFCCs.csv')
df.info()
```

```python
df['Family'] = df['Family'].apply(lambda x: 1 if x=='Dendrobatidae' else 0)
X = df.drop(['Family', 'Genus', 'Species', 'RecordID'], axis=1)
y = df['Family']
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state = 31)
print(X_train.shape)
print(X_test.shape)
```

```python
rf1 = ensemble.RandomForestClassifier(n_estimators=10, random_state=42)
rf1.fit(X_train, y_train)
y_pred = rf1.predict(X_test)
print('F1 score for test set: {:.2f}'.format(metrics.f1_score(y_test, y_pred)))
```

```python
estimators = [
    ('rf', ensemble.RandomForestClassifier(n_estimators=10, random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=11)),
    ('GNB', GaussianNB())
    ]
stack = ensemble.StackingClassifier(estimators=estimators, final_estimator=linear_model.LogisticRegression())
stack.fit(X_train, y_train)
y_pred_stack = stack.predict(X_test)
print('F1 score for test set: {:.2f}'.format(metrics.f1_score(y_test, y_pred_stack)))
```
